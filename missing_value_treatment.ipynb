#### Missing value treatment

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\pandas\missing.csv')
```

```python
df
```

```python
# Numerical - number
# with outlier : median
# without outlier : mean

# Categorical - string - use mode to fill missing
```

```python
df['dept'] = df['dept'].fillna(df['dept'].mode()[0])
```

```python
df
```

```python
df['salary'].plot(kind='box')
```

```python
df['Age'].plot(kind='box')
```

```python
df['salary'].mean()
```

```python
df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python
df
```

```python
df['Age'].median()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```
