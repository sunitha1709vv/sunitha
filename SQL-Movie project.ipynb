# overdue rentals

------------------------
In our rental table we have the dates when a movie has been rented and the date when it has been returned. Therefore, we can calculate how many days the customer has held the movie and compare this with the maximum rental duration for the movie , in order to see if they have held it more than "allowed" -> was it overdue?

Let's first query this information, before making the comparisons. For the rental and return dates we use the srtftime() function to show just the year-month-day i.e. without the exact time. For the holding period we use the JULIANDAY() function to calculate the differences between the two dates, because this function calculates fractions of days, which is what we are after in this case. ROUND() is used to round the difference to 2 decimal places and WHERE r.return_date IS NOT NULL is used because when the return_date is NULL, that means the movie has not been returned yet.


```python
pd.read_sql("""

SELECT
    r.rental_id AS Rental_ID
    ,f.title AS Title
    ,strftime('%Y-%m-%d', r.rental_date) AS Rent
    ,strftime('%Y-%m-%d', r.return_date) AS Return
    ,ROUND(JULIANDAY(r.return_date) - JULIANDAY(r.rental_date), 2) AS HoldPeriod
    ,f.rental_duration AS RentalDuration
FROM rental AS r
JOIN inventory AS i ON r.inventory_id = i.inventory_id
JOIN film AS f ON i.film_id = f.film_id
WHERE r.return_date IS NOT NULL;

""", connection)
```

#now let us calculate hte difference between the HoldPeriod and RentalDuration and retreive only this rows which show overdue rentals and query the corrsponding customers
(to quiery the corresponding customers we will join the customer table as well)

```python
pd.read_sql("""

SELECT
    r.rental_id AS Rental_ID
    ,c.last_name || ', ' || c.first_name AS Name
    ,c.email
    ,f.title As Film
    ,ROUND(JULIANDAY(r.return_date) - JULIANDAY(r.rental_date), 2) - f.rental_duration AS Overdue_Days
FROM rental AS r
JOIN inventory AS i ON r.inventory_id = i.inventory_id
JOIN film AS f ON i.film_id = f.film_id
JOIN customer AS c ON r.customer_id = c.customer_id
WHERE r.return_date IS NOT NULL
AND Overdue_Days > 0;

""", connection)
```

now let us see which customers have had the most overdue rentals .for this once agin we group by the customer name and use a count() function on the group data .Note that the Overdue_Days in this case have been used as a filter in the where clause .

```python
pd.read_sql("""

SELECT
    c.customer_id AS Customer_ID
    ,c.last_name || ', ' || c.first_name AS Name
    ,c.email
    ,COUNT(*) AS Overdue_Count
FROM rental AS r
JOIN inventory AS i ON r.inventory_id = i.inventory_id
JOIN film AS f ON i.film_id = f.film_id
JOIN customer AS c ON r.customer_id = c.customer_id
WHERE r.return_date IS NOT NULL
AND ROUND(JULIANDAY(r.return_date) - JULIANDAY(r.rental_date), 2) - f.rental_duration > 0
GROUP BY Name
ORDER BY Overdue_Count DESC, Name ASC;

""", connection)
```

#the above query retreives customer information such as customer id ,Full name ,email,and count of overdue rental for each customer.the connection variables represnts a valid connection to the sqlite dataabse to execute the quieries successfully.

print the total count for each customer id 


```python
pd.read_sql("""

SELECT 
    customer.customer_id AS Customer_ID
    ,COUNT(*) AS Total_Count
FROM rental
JOIN customer ON rental.customer_id = customer.customer_id
WHERE return_date IS NOT NULL
GROUP BY rental.customer_id
ORDER BY Total_Count DESC;

""", connection)
```

lets calculate the percenage of overdue rentals .it is taking two same tables from the above use the subqueries and joining them together on the customer_id to reteives the column from the resuting join table.

```python
pd.read_sql("""

SELECT
    Overdue_Table.Name
    ,Overdue_Table.email
    ,ROUND(Overdue_Table.Overdue_Count * 1.0 / Total_Table.Total_Count, 2) AS PercentOverdue
    

-- FIRST TABLE
FROM (SELECT 
    customer.customer_id AS Customer_ID
    ,COUNT(*) AS Total_Count
FROM rental
JOIN customer ON rental.customer_id = customer.customer_id
WHERE return_date IS NOT NULL
GROUP BY rental.customer_id
ORDER BY Total_Count DESC) AS Total_Table

--SECOND TABLE
JOIN 

(SELECT
    c.customer_id AS Customer_ID
    ,c.last_name || ', ' || c.first_name AS Name
    ,c.email
    ,COUNT(*) AS Overdue_Count
FROM rental AS r
JOIN inventory AS i ON r.inventory_id = i.inventory_id
JOIN film AS f ON i.film_id = f.film_id
JOIN customer AS c ON r.customer_id = c.customer_id
WHERE r.return_date IS NOT NULL
AND ROUND(JULIANDAY(r.return_date) - JULIANDAY(r.rental_date), 2) - f.rental_duration > 0
GROUP BY Name
ORDER BY Overdue_Count DESC, Name ASC) AS Overdue_Table


ON Total_Table.Customer_ID = Overdue_Table.Customer_ID
ORDER BY PercentOverdue DESC;

""", connection)
```

ROUND(Overdue_Table.Overdue_Count * 1.0 / Total_Table.Total_Count, 2) AS PercentOverdue-  this calculates the percentage of overdue rentals for each customer .it first divides the count of overdue rentals (Overdue_count) from the overdue_table by the total count of rental (total_count).from the total_table.the resuts is then rounded to two decimal places and named as PercentOverdue.

the quiery uses to two subquiery denoted as Total_Table and Overdue_Table to calculate the total count of rentals and the ccount of overdue rentals respectively .

the total table subquiery calculates te total count of rental for each customer who has returned the films. it joins the rental and custemer table filters out te rental with a non null return date and overdue rentals ,groups the resut by customer id and calculates the  total count of rentals of each customers.the results is sorted in total count.

the Overdue_Table subquiery calculates the count of overdue rentals for each customer 
The joint operation combines the two tabes based on customer_id .



```python
#let us check if there  has been extra payment or overdue taxes in cases when a movie has a return after the maximum rental duration 
#Hint here we w ill use leftjoin because we want to rteurn all payments even if  for some reasons they are ot assocuiated with the rentals.
#let us group by rental to see if there has been multiple payment for one rental 
pd.read_sql("""

SELECT 
    r.rental_id
    ,COUNT(*) AS Count_Payments
FROM payment AS p
LEFT JOIN rental AS r
ON p.rental_id = r.rental_id
GROUP BY p.rental_id
ORDER BY Count_Payments DESC
LIMIT 10;

""", connection)
```
